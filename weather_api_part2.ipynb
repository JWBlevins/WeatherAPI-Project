```python
# current weather

current_url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}" 
current_resp = requests.get(current_url)
current_resp.raise_for_status()
    
current = current_resp.json()["current"]
location = current_resp.json()["location"]

print("City:", location["name"], location["region"])
print("Temp (F):", current["temp_f"])
print("Condition:", current["condition"]["text"])
```

```python
# forecast weather
forecast_url = f"http://api.weatherapi.com/v1/forecast.json?key={api_key}&q={city}&days=3&aqi=no&alerts=no"

forecast_resp = requests.get(forecast_url)
forecast_resp.raise_for_status()

forecast_days = forecast_resp.json()["forecast"]["forecastday"]

for day in forecast_days:
    print(
        day["date"],
        "High:", day["day"]["maxtemp_f"], "F",
        "Low:", day["day"]["mintemp_f"], "F",
        "-", day["day"]["condition"]["text"]
    )
```

```python
# 3) --- Astronomy for today (sunrise/sunset, moon)

from datetime import date
import requests


today_str = date.today().strftime("%Y-%m-%d")
astro_url = f"http://api.weatherapi.com/v1/astronomy.json?key={api_key}&q={city}&dt={today_str}"

astro_resp = requests.get(astro_url)
astro_resp.raise_for_status()

astro = astro_resp.json()["astronomy"]["astro"]

astronomy_summary = {
    "sunrise": astro["sunrise"],
    "sunset": astro["sunset"],
    "moon_phase": astro["moon_phase"],
    "moon_illumination": astro["moon_illumination"]
}

astronomy_summary  
```
